```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def test_integ(x, a):
        
        I = sp.integrate.quad(lambda y: np.sin(y), 0, x)[0]
        
        return I
    
x = np.linspace(0, 2*np.pi, 10000)
xx = np.zeros(len(x))
for i in range(len(x) - 1):
    
    xx[i] = test_integ(x[i], 1)
    
xx
plt.figure()
plt.plot(x, xx)
plt.plot(x, -np.cos(x) + 1, ls = '--', c = 'r')
plt.show()
```
